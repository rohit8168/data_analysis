```python
print("no of female in data is:",(data['Gender']=='F').sum())
print("no of male in data is:",(data['Gender']=='M').sum())
total_human=len(data['User_ID'])
per=((data['Gender']=='M').sum())/total_human
print('percentage of woman ',per)
dat=[(data['Gender']=='M').sum(),(data['Gender']=='F').sum()]
label=['M','F']
plt.pie(dat,labels=label,autopct='%2.2f%%')
plt.show()
```

```python
print("no of married",(data['Marital_Status']==1).sum())
print("no of non-married",(data['Marital_Status']==0).sum())
dat=[(data['Marital_Status']==1).sum(),(data['Marital_Status']==0).sum()]

plt.pie(dat,labels=['married','non-married'],autopct='%2.2f%%')
plt.show()
```
