```python
df=pd.read_csv(r"C:\ml\da\IRIS.csv")
df.head()
```

```python
plt.subplot(2,2,1)
sns.distplot(df['sepal_length'])
plt.subplot(2,2,2)
sns.distplot(df['sepal_width'])
print("--"*35)
plt.subplot(2,2,3)
sns.distplot(df['petal_length'])
plt.subplot(2,2,4)
sns.distplot(df['petal_width'])
```

```python
mean=df['sepal_width'].mean()
std=df['sepal_width'].std()
print(mean,std)
```

```python
high=mean+(3*std)
low=mean-(3*std)
print(high,low)
```

```python
df=df[(df['sepal_width']>low)&(df['sepal_width']<high)]
df
```

#making z score


```python
df['z-score']=(df['sepal_width']-mean)/std
df[(df['z-score']>-3)&(df['z-score']<3)]
```

```python
q1=df3['weight'].quantile(0.25)
q3=df3['weight'].quantile(0.75)
print(q1,q3)
```

```python
IQR=q3-q1
IQR
```

```python
df3.describe()
```

```python
#ouliers
df3[(df3['weight']<mini)|(df3['weight']>maxi)]
```
