```python
import matplotlib as mp
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,360*2,30*5)

siny=np.sin(np.deg2rad(x))
cosy=np.cos(np.deg2rad(x))

plt.plot(x, siny, '*',label="sin")
plt.plot(x, cosy, '.',label='cos')
plt.title('Cos vs sin')
plt.legend(loc='upper right')
plt.show()
```

```python
#multiple
X=np.linspace(0,10,50)
fiG, ax=plt.subplots(2,2)
Y=np.sin(X)
ax[0,0].plot(X,Y)
ax[0,0].set_title('sin_wave')
Y=np.cos(X)
ax[0,1].plot(X,Y)
ax[0,1].set_title("cos")
Y=np.tan(X)
ax[1,0].plot(X,Y)
ax[1,0].set_title("tan")
Y=np.random.rand(1000)
ax[1,1].hist(Y)
ax[1,1].set_title("random")
plt.tight_layout()
plt.show()
```
